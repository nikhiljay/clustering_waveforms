```python
#This function will work for URL calls for <40,000 events at a time

def downCat(catalog,minlat,maxlat,minlon,maxlon,start,end,minmag,maxmag):
    if(catalog=="ISC"):
        URL="http://isc-mirror.iris.washington.edu/fdsnws/event/1/query?starttime="+\
        start+"&endtime="+end+"&minlatitude="+minlat+"&maxlatitude="+maxlat+"&minlongitude="+minlon+"&maxlongitude="+\
        maxlon+"&minmagnitude="+minmag+"&maxmagnitude="+maxmag+""
        
    elif(catalog=="USGS"):
        URL="https://earthquake.usgs.gov/fdsnws/event/1/query?format=xml&starttime="+\
        start+"&endtime="+end+"&minmagnitude="+minmag+"&maxmagnitude="+maxmag+"&minlatitude="+\
        minlat+"&maxlatitude="+maxlat+"&minlongitude="+minlon+"&maxlongitude="+maxlon+""
    
    eventCatalog=read_events(URL)
    
    return eventCatalog
```

```python
#Uses downCat to break up a download into multiple downloads
#useful for getting around maximum event limits
def downCats(catalog,minlat,maxlat,minlon,maxlon,start,end,minmag,maxmag):
    
    masterCat=Catalog()
    
    magstep=1.0
    maglist=np.arange(minmag,maxmag+0.5*magstep,magstep)
    for i in range(len(maglist)-1):
        if i>0:
            minM=str(maglist[i]+0.0001) #to avoid duplicates with intersecting magnitude thresholds
        else:
            minM=str(maglist[i])
        maxM=str(maglist[i+1])
        
        print("Downloading set: "+minM,maxM)
        
        try:
            cat=downCat(catalog,minlat,maxlat,minlon,maxlon,start,end,minM,maxM)
            masterCat.extend(cat)
            
        except requests.HTTPError:
            print("No events found in range")
        
        print("Download loop #"+str(i+1)+" complete")
        
    return masterCat
```

```python
#Search for Mainshock

minmag=6.6
maxmag=10

mainCat=downCats("USGS",minlat,maxlat,minlon,maxlon,start,end,minmag,maxmag)
```

```python
#Aftershock Time Sequence
#Figure 2 - Magnitude vs Time with lines

#Plot Magnitude vs Time // Timeline of Earthquake Sequence
def magnitudeTime(dataframe,limit):
    
    #All magnitudes greater than or equal to the limit will be plotted
    
    plt.figure(figsize=(16,12))
    ax=plt.subplot()

    ax.set_title("Magnitude vs Time")
    ax.set_xlabel("Date (UTC)")
    ax.set_ylabel("Magnitude")

    for i in range(len(dataframe)):
        time=dataframe.datetime[i]
        mag=dataframe.magnitude[i]
        
        #Create a python datetime object
        pTime=time.datetime
        
        if(mag>=limit):
            plt.plot(pTime,mag,c="r",marker="v",markersize=(6))
            plt.vlines(x=pTime,ymin=limit,ymax=mag,colors="k",linewidth=1)
            
            

    #plt.savefig("MagVsTime") 
    plt.grid(axis="both")
    plt.margins(y=0.005)
    plt.show()
```

```python
def exponfunc(x, a, b, c):
    return a * np.exp(-b * x) + c
```

```python
#Figure 4 - Earthquake Recurrence Time
#Histogram of Earthquake magnitudes between 1-2 day intervals
#We want the scale of the exponential decay


#start must be the date of the mainshock, end can be any date as long as there is data
def earthRtime(dataframe,interval,start,end):
    
    dates=pd.date_range(start,end)
    
    dayCount=[]
    for date in dates:
        count=0
        for i in range(len(dataframe)):
            #If there is an event on this day...
            if date==dataframe.pDate[i]:
                #Add one to the count list
                count+=1
        dayCount.append(count)
    
    intervalCount=[]
    #Start at day 1 and interval 0
    day=1
    i=0
    currentCount=0
    
    for count in dayCount:
        currentCount+=count
        day=day+1
        
        #print("day:"+ str(day))
        #print("count:"+str(count))
        #print("currntCount:"+str(currentCount))
        
        #If day goes beyond current interval
        if(day>interval):
            #New interval
            i=i+1
            day=1
            intervalCount.append(currentCount)
            currentCount=0
    
    #Create the graph
    plt.figure(figsize=(16,12))
    ax=plt.subplot()
    
    ax.set_xlabel("Interval of Days after Mainshock")
    ax.set_ylabel("Number of earthquakes in interval")
    
    bins=np.arange(len(intervalCount))
    
    norm=np.array(intervalCount)/sum(intervalCount)
    
    plt.bar(bins,norm,facecolor = '#2ab0ff', edgecolor='#169acf',align="edge",width=1)
    
    popt,pcov=optimize.curve_fit(exponfunc,bins,norm)
    
    plt.plot(bins,exponfunc(bins,*popt),"k-",label='y=a*e^(-b*x)+c where fit: a=%5.3f, b=%5.3f, c=%5.3f' % tuple(popt))
    
    #get decay value
    decay=popt[1]
    
    #Create bar labels/ticks
    day1=1
    ticks=[]
    for i in range(len(intervalCount)):
        day2=day1+interval-1
        ticks.append(str(day1)+"-"+str(day2))
        day1=day2+1
      
    #Set xticks
    plt.xticks(bins,ticks)
    
    #Remove extra white space
    plt.xlim([0,bins.size])
    
    plt.legend()
    plt.show()
    
    print("Total Bins (Length of intervalCount): "+str(len(intervalCount)))
    for i in range(len(intervalCount)):
        print("Bin "+ticks[i]+": "+str(intervalCount[i]))
        
    #return decay constant, save for later
    return decay
```

```python
#Frequency - Magnitude
#WORK IN PROGRESS :(
#Needs to find and return the magnitude of completness

def freqmag(dataframe,dmag=0.2):
    mags=dataframe.magnitude
    
    ms=np.arange(0,10,dmag)
    ms=list(ms)
    
    hist,bin_edges=np.histogram(mags,bins=ms)
    hist=list(hist)
    
    #Remove magnitude 0
    ms.pop(0)
    
    counts=[]
    for i in range(len(hist)):
        total=0
        for j in range(i,len(hist)):
            total+=hist[j]
        counts.append(total)
    
    #Remove zeros
    zeros=counts.count(0)
    
    for i in range(zeros):
        pos=counts.index(0)
        ms.pop(pos)
        counts.pop(pos)
        
    #Take log of counts
    values=np.log10(counts)
    
    #Plot ms as Magnitudes and Values as log of the counts
    plt.figure(figsize=(16,12))
    ax=plt.subplot()

    ax.set_title("Aftershock Frequency-Magnitude")
    ax.set_xlabel("M")
    ax.set_ylabel("10^N(>M)")
            
    #Create a function of the graph
    coef = np.polyfit(ms,values,6)
    function = np.poly1d(coef)
    
    plt.plot(ms,values,"bo",label="Data")
    #plt.plot(ms,values,"b")
    
    plt.plot(ms,function(ms),"--k",label="Best Fit Line")
    
    
    #Take 1st derivative of function
    
    #SciPy's derivative function takes the derivative at a single point
    #Idea: loop through each ms value and take derivative of each
    mprime=[]
    for m in ms:
        mprime.append(derivative(function,m,dx=0.1))
    
    plt.plot(ms,mprime,"--r",label="1st Derivative of best fit")
    
    #Paper states that the maximum value of the first derivative
    #also, add 0.2 to compensate for underestimate of MAXC method
    #print("Estimated Mc value = "+str(x+0.2))
    
    bValue=0
    mcValue=0
    #Return magnitude of completness and b-value

    ax.grid()
    plt.legend()
    plt.show()
    return bValue,mcValue,ms,values
```

```python
def rms(x):
    return np.sqrt(x.dot(x)/x.size)
```

```python
magcheck=np.arange(1,5+0.5*dmag,dmag)
magmax=5.5
magall=np.array(mag)
# print(magall)
rmsall=[]
rmsmag=[]
rr2all=[]
for i,m in enumerate(magcheck):
#     print(magcheck[i])
    idx1=np.where(magall <= magcheck[i])[0]
    idx2=np.where((magall > magcheck[i]) & (magall <= magmax))[0]
#     print(idx1,idx2)
    mag1=magall[idx1]
    val1=values[idx1]
    mag2=magall[idx2]
    val2=values[idx2]
    
    coef = np.polyfit(mag1,val1,1)
    function = np.poly1d(coef)
    pred1=function(mag1)
    rms1=rms(np.abs(pred1-val1))
    rr1=scipy.stats.pearsonr(pred1,val1)[0]
    coef = np.polyfit(mag2,val2,1)
    function = np.poly1d(coef)
    pred2=function(mag2)
    rms2=rms(np.abs(pred2-val2))
    rr2=scipy.stats.pearsonr(pred2,val2)[0]
    rr2all.append(rr2)
    rmsmag.append(magcheck[i])
    rmsall.append(rms1+rms2)
    
#     plt.figure(figsize=(8,3))
#     plt.plot(mag1,val1,'o')
#     plt.plot(mag2,val2,'o')
#     plt.xlim([min(magall),magmax])
#     plt.text(0.7,3.8,'rms1=%6.4f, rr1=%6.4f'%(rms1,rr1))
#     plt.text(0.7,3.6,'rms2=%6.4f, rr2=%6.4f'%(rms2,rr2))
#     plt.text(0.7,3.4,'rms_all=%6.4f'%(rms1+rms2))
#     plt.show()
```
